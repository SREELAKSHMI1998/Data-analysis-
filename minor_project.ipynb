```python
df1=df[['budget', 'revenue', 'original_title',
       'cast','runtime', 'genres', 'production_companies','release_year', 'budget_adj',
        'revenue_adj']]
```

```python
df1=df1.dropna()
```

```python
df1.info()
```

```python
#QUESTION NO: 1
# movies with the third lowest and third highest budget?
```

```python
third_highest_budget=df1.sort_values(by='budget_adj',ascending=False)
third_lowest_budget=df1.sort_values(by='budget_adj')
print("third highest budget film:",third_highest_budget.iloc[2,2])
print("third lowest budget film:",third_lowest_budget.iloc[2,2])
```

```python
third_highest_budget[['original_title','budget_adj']].head(3) #order of budget is in descending order
```

```python
third_lowest_budget[['original_title','budget_adj']].head(5) #order of budget is in ascending order
```

```python
#QUESTION NO: 2
#Which are the movies with most and least earned revenue?
```

```python
highest_revenue=df1.sort_values(by='revenue_adj',ascending=False)
lowest_revenue=df1.sort_values(by='revenue_adj')
print("highest revenue film:",highest_revenue.iloc[0,2])
print("lowest revenue film:",lowest_revenue.iloc[0,2])
```

```python
#QUESTION NO: 3
#What is the average runtime of movies in the year 2006?
```

```python
average_run_time=df1.loc[(df1['release_year'] == 2006)]
avg=average_run_time['runtime'].mean()
print("average runtime of movies released in the year 2006 is:",avg)
```

```python
#question no:4
# What is the average number of words in movie titles between the year 2000-2005?
movies=df1.loc[(df1['release_year']>=2000 ) & (df1['release_year']<=2005)]
number_of_words={}
for i in range(movies.shape[0]):
    words=str(movies['original_title'].values[i]).split(' ')
    for j in words:
        try:
            count=number_of_words[j]
            number_of_words[j]=count+1
        except:
            number_of_words[j]=1
            
number_of_words
```

```python
total_no_words=sum(number_of_words.values())
print("total number of words used :",total_no_words)
distinct_no_words=len(number_of_words.keys())
print("total number of distinct words :",distinct_no_words)
```

```python
average_number_of_words=total_no_words/distinct_no_words
print("the average number of words in movie titles between the year 2000-2005 :",average_number_of_words)
```
